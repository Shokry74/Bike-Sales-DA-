## Data Cleaning & Data Prepration:

```python
# delete the (ID,Childern,Home Owner) columns 
data.drop(["ID","Children","Home Owner"],axis = "columns",inplace=True)
data.head(10)
```

```python
data["Marital Status"] = data["Marital Status"].replace({"M":"Married","S":"Single"})
data.head(10)
```

```python
data["Gender"] = data["Gender"].replace({"F":"Female","M":"Male"})  
data.head(10)
```

```python
# Function to create age brackets
def age_bracket(age):
    if age < 18:
        return 'Child'
    elif 18 <= age < 30:
        return 'Young Adult'
    elif 30 <= age < 45:
        return 'Adult'
    elif 45 <= age < 60:
        return 'Middle Aged'
    else:
        return 'Senior'

# Apply the function to create a new column
data['Age Bracket'] = data['Age'].apply(age_bracket)
data.head(10)
```

# Data Visualization:

```python
summary = data.groupby(["Age Bracket", "Purchased Bike"]).size().unstack(fill_value=0)
summary.plot(kind="line", stacked=False, figsize=(8, 6), color=["skyblue", "salmon"])

# Adding labels and title
plt.title("Purchased Bike by Age Bracket", fontsize=14)
plt.xlabel("Age Bracket", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Purchased Bike", loc="upper right")
plt.show()
```

```python
summary = data.groupby(["Gender", "Purchased Bike"]).size().unstack(fill_value=0)
summary.plot(kind="bar", stacked=False, figsize=(8, 6), color=["skyblue", "salmon"])

# Adding labels and title
plt.title("Purchased Bike by Gender", fontsize=14)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Purchased Bike", loc="upper right")
plt.show()
```

```python
# Grouping data
grouped = data.groupby(["Gender", "Purchased Bike"])["Income"].mean().unstack()

# Plotting
fig, ax = plt.subplots(figsize=(8, 6))
width = 0.35  # Width of bars
x = np.arange(len(grouped.index))  # Gender categories

ax.bar(x - width/2, grouped["Yes"], width, label="Purchased Bike: Yes", color="skyblue")
ax.bar(x + width/2, grouped["No"], width, label="Purchased Bike: No", color="salmon")

# Adding labels and title
ax.set_xlabel("Gender", fontsize=12)
ax.set_ylabel("Average Income", fontsize=12)
ax.set_title("Average Income by Gender and Purchased Bike", fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(grouped.index)
ax.legend()

# Show plot
plt.tight_layout()
plt.show()
```
